```python
import pandas as pd 
import seaborn as sns
```

```python
df=pd.read_csv("data.csv")
df.corr(numeric_only=True)
```

```python
df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})
df['Partner'] = df['Partner'].map({'Yes': 1, 'No': 0})
selected_col=['Churn','Partner','tenure','MonthlyCharges','MonthlyCharges']
print(df[selected_col].corr())
sns.heatmap(df[selected_col].corr(),annot=True, cmap='coolwarm', fmt=".2f")
```
